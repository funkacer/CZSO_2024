```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
def add_labels(line):
    x,y=line.get_data()
    labels=map(':'.join,zip(map(lambda s: '%g'%s,x),map(lambda s: '%g'%s,y)))
    list(map(plt.text,x,y,labels))
```

```python
x = np.arange(0, 10)
y = np.exp2(x)
line,= plt.plot(x, y, marker = "x")
#print(line.get_data())
#x,y=line.get_data()
#labels=map(','.join,zip(map(lambda s: '%g'%s,x),map(lambda s: '%g'%s,y)))
#print(list(map(plt.text,x,y,labels)))
add_labels(line)
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.show()
```

```python
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

bins = 11
x = np.linspace(-5, 5, 1000)
y = np.random.normal(size=len(x))

hist, bin_edges = np.histogram(y, density=True, bins = bins)

print(len(hist), len(bin_edges))
```

```python
bin_width = (np.max(bin_edges) - np.min(bin_edges)) / len(hist)
```

```python
fig = plt.figure(figsize=(4, 3), facecolor='w')
#plt.hist(y, alpha = 0.5)
plt.bar(bin_edges[:bins]+bin_width/2, hist, color = "g", alpha = 0.5, width = bin_width-0.1)
plt.title("Hist using adjusted bin_edges", fontsize=10)
data = io.BytesIO()
plt.savefig(data, bbox_inches='tight')
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
plt.close(fig)
```

```python
display.display(display.Markdown(F"""![{alt}]({image})"""))
```
