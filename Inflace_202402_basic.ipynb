```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as display
import io
import base64
```

```python
with open("_data/202402/CEN08B3_mesicni_data.txt", encoding = "utf-16") as f:
    line = f.readline()
    row = 0
    data = []
    row_indexes = []
    while line != "":
        if row == 0:
            # prvni zaznam je 'Oddíl ECOICOP', coz je nas index
            columns = line.replace("\n", "").replace('"', "").split("\t")[1:]
        else:
            processed_line = line.replace("\n", "").replace(",", ".").split("\t")
            row_indexes.append(processed_line[0])
            data.append(processed_line[1:])
        line = f.readline()
        row += 1

print(columns)
print(data)
```

```python
for i in range(len(data)):
    assert len(columns) == len(data[i]), "Pozor, nesedi pocet sloupcu"
```

```python
df = pd.DataFrame(data, index = row_indexes)
df.columns = columns
```

# Teď zkusím spočítat hodnoty z webu

Při vyjadřování **míry inflace** pomocí **indexu spotřebitelských cen** jsou často uváděna různá čísla, která, i když rozdílná, jsou správná. Je třeba uvést jednoznačně období, za které je míra inflace uváděna, a základ, k němuž se vymezené období porovnává. Nejčastěji se používají:

**Míra inflace vyjádřená přírůstkem průměrného ročního indexu spotřebitelských cen** vyjadřuje procentní změnu průměrné cenové hladiny za 12 posledních měsíců proti průměru 12 předchozích měsíců.

>únor 2024: **8,2 %**

Datum zveřejnění: 11.03.2024

```python
display.display(display.Markdown("""### Vyberu posledních ${\color{green}12}$ ${\color{green}měsíců}$ pomocí ${\color{green}loc}$."""))
df1.loc['2023-03' : '2024-02']
```

```python
display.display(display.Markdown("""### Vyberu posledních ${\color{green}12}$ ${\color{green}měsíců}$ a jen ${\color{green}'Úhrn'}$."""))
df1.loc['2023-03' : '2024-02', 'Úhrn']
```

```python
display.display(display.Markdown("""### Udělám ${\color{green}průměr}$ za posledních ${\color{green}12}$ ${\color{green}měsíců}$:"""))
df1.loc['2023-03' : '2024-02', 'Úhrn'].mean()
```

```python
display.display(display.Markdown("""### A odečtu ${\color{green}100}$:"""))
df1.loc['2023-03' : '2024-02', 'Úhrn'].mean() - 100
```

**Průměrná roční míra inflace –** jedná se o hodnotu téhož ukazatele v prosinci daného roku.
>rok 2023: **10,7 %**

```python
display.display(display.Markdown("""### Vyberu rok ${\color{green}2023}$ všech ${\color{green}12}$ ${\color{green}měsíců}$."""))
df1.loc['2023-01' : '2023-12']
```

```python
display.display(display.Markdown("""### Vyberu rok ${\color{green}2023}$  všech ${\color{green}12}$ ${\color{green}měsíců}$ a jen ${\color{green}'Úhrn'}$."""))
df1.loc['2023-01' : '2023-12', 'Úhrn']
```

```python
display.display(display.Markdown("""### Udělám ${\color{green}průměr}$ za rok ${\color{green}2023}$  všech ${\color{green}12}$ ${\color{green}měsíců}$ a jen ${\color{green}'Úhrn'}$."""))
df1.loc['2023-01' : '2023-12', 'Úhrn'].mean()
```

```python
display.display(display.Markdown("""### A odečtu ${\color{green}100}$:"""))
df1.loc['2023-01' : '2023-12', 'Úhrn'].mean() - 100
```

**Míra inflace vyjádřená přírůstkem indexu spotřebitelských cen ke stejnému měsíci předchozího roku** vyjadřuje procentní změnu cenové hladiny ve vykazovaném měsíci daného roku proti stejnému měsíci předchozího roku.

>únor 2024: **2,0 %**

Datum zveřejnění: 11.03.2024

```python
display.display(display.Markdown("""### Vyberu měsíc ${\color{green}únor}$ a jen ${\color{green}'Úhrn'}$ a odečtu ${\color{green}100}$:"""))
df1.loc['2024-02', 'Úhrn'] - 100
```

**Míra inflace vyjádřená přírůstkem indexu spotřebitelských cen k předchozímu měsíci** vyjadřuje procentní změnu cenové hladiny sledovaného měsíce proti předchozímu měsíci.
